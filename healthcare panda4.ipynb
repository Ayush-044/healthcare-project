```python
import pandas as pd
```

```python
df= pd.read_csv("healthcare_dataset4.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
```

```python
df["Date of Admission"]=pd.to_datetime(df["Date of Admission"],errors= "coerce")
df.head()
```

```python
df["Discharge Date"]=pd.to_datetime(df["Discharge Date"],errors= "coerce")
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.tail()
```

```python
df["Stay Duration"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days                                 # Add new column Stay duration 
```

```python
df.head()
```

```python
df["Name"] = df["Name"].str.title()                                           # Convert names to proper case (title case)
df.head()
```

```python
df["Gender"] = (df["Gender"]
    .str.strip()                                                        # remove spaces
    .map({"Male": "M", "Female": "F"})                                  # convert
)
df.head()
```

```python
print(df["Gender"].unique())
```

```python
df.tail()
```

```python
df.info()
```

```python
df.to_csv("cleaned_healthcare.csv",index= False)
```
